```python
import pandas as pd
import numpy as np
```

```python

df1 = pd.read_excel("../Donnees/analyse de terre.XLSX")
df2 = pd.read_excel("../Donnees/BiomasseMoleculaireMicrobienne.xlsx")
df3 = pd.read_csv("../Donnees/DataVdt_N.csv")
df4 = pd.read_excel("../Donnees/TableauxdonnéesParcelle.xlsx")

id_col_1 = 'Réf_échantillon'
id_col_2 = 'code echantillon'
id_col_3 = df3.columns[0]
id_col_4 = 'Parcelle'

df_merged = df1.merge(df2, left_on=id_col_1, right_on=id_col_2)
df_merged = df_merged.merge(df3, left_on=id_col_1, right_on=id_col_3)
df_merged = df_merged.merge(df4, left_on=id_col_1, right_on=id_col_4)

result_df = pd.DataFrame()
result_df[id_col_1] = df_merged[id_col_1]
result_df = result_df.join(df_merged[[
    ' Matière organique ', 
    ' Azote total', 
    'Moy E1_E2',  
    'somme_espece_biodiv', 
    'Ganivelles', 
    'Emplacement', 
    'Enfantsparm2'
]])

result_df.to_csv("../Donnees/excel_combiné.csv", index=False)

result_df.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Supposons que votre DataFrame s'appelle data_modif

# Supprimer les espaces autour des noms de colonnes
data_modif.columns = data_modif.columns.str.strip().str.lower()  # Convertir en minuscules pour uniformité

# Définir vos colonnes à normaliser
colonnes_a_normaliser = ['argile', 'limon fin', 'limon grossier', 'sable fin', 'sable_grossier', 
                         'ph_eau', 'carbonates_totaux', 'carb_organique', 'matière organique', 
                         'azote total', 'c/n -', 'p2o5 olsen', 'cu dtpa mg/kg', 'zn dtpa mg/kg', 
                         'ni dtpa mg/kg', 'cu total er (mg/kg)', 'ni total er (mg/kg)', 
                         'pb total er (mg/kg)', 'zn total er (mg/kg)']

# Remplacer les virgules par des points dans les colonnes à normaliser
data_modif[colonnes_a_normaliser] = data_modif[colonnes_a_normaliser].replace(',', '.', regex=True)

# Convertir les colonnes en format numérique
data_modif[colonnes_a_normaliser] = data_modif[colonnes_a_normaliser].apply(pd.to_numeric, errors='coerce')

# Normaliser les données
scaler = StandardScaler()
data_normalise = scaler.fit_transform(data_modif[colonnes_a_normaliser])

# Afficher les données normalisées
print(data_normalise)
```

```python
# Supprimer les lignes avec des valeurs NaN
data_modif = data_modif.dropna(subset=colonnes_a_normaliser)
```

```python

data_modif[colonnes_a_normaliser] = data_modif[colonnes_a_normaliser].fillna(data_modif[colonnes_a_normaliser].mean())
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
from sklearn.preprocessing import StandardScaler

# Normaliser les données
scaler = StandardScaler()
data_normalise = scaler.fit_transform(data_modif[colonnes_a_normaliser])
```

```python
# Définir le nombre de clusters
n_clusters = 3  # Par exemple, 3 clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
clusters = kmeans.fit_predict(data_normalise)

# Ajouter les étiquettes de cluster au DataFrame
data_modif['cluster'] = clusters
```

```python
silhouette_avg = silhouette_score(data_normalise, clusters)
print(f'Indice de Silhouette pour {n_clusters} clusters: {silhouette_avg:.3f}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualisation des clusters (vous pouvez choisir d'autres paires de caractéristiques)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_modif['matière organique'], y=data_modif['azote total'], hue=data_modif['cluster'], palette='viridis', s=100)
plt.title('Clusters de Sols avec K-Means')
plt.xlabel('matière organique')
plt.ylabel('azote total')
plt.legend(title='Cluster')
plt.show()
```

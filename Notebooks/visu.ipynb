```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

result_df = pd.read_csv("../Donnees/excel_combiné.csv")

print(result_df[['somme_espece_biodiv', 'Ganivelles']].head())

result_df['Ganivelles'] = result_df['Ganivelles'].astype(str)

plt.figure(figsize=(10, 6))
sns.barplot(x='Ganivelles', y='somme_espece_biodiv', data=result_df, estimator=sum, ci=None)

plt.title('Somme de la biodiversité en fonction de la présence de ganivelles')
plt.xlabel('Ganivelles')
plt.ylabel('Somme de l\'espèce biodiversité')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger le CSV
result_df = pd.read_csv("../Donnees/excel_combiné.csv")

# Vérifier les données pour les colonnes d'intérêt
print(result_df[['somme_espece_biodiv', 'Emplacement']].head())

# Créer un graphique à barres pour la somme de la biodiversité par emplacement
plt.figure(figsize=(12, 6))
sns.barplot(x='Emplacement', y='somme_espece_biodiv', data=result_df, estimator=sum, ci=None)

# Personnaliser le graphique
plt.title('Somme de la biodiversité par Emplacement')
plt.xlabel('Emplacement')
plt.ylabel('Somme de l\'espèce biodiversité')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Afficher le graphique
plt.tight_layout()
plt.show()
```

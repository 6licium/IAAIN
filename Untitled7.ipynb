```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
import torch
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from datasets import Dataset
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
import json
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns

# S'assurer que tout est sur le bon pÃ©riphÃ©rique (GPU ou CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 1. Parser les fichiers WebAnno TSV
def parse_webanno_tsv(file_path):
    with open(file_path, encoding='utf-8') as f:
        lines = f.readlines()

    texts, labels = [], []
    current_tokens, current_labels = [], []

    for line in lines:
        if line.startswith("#") or line.strip() == "":
            if current_tokens:
                full_text = " ".join(current_tokens)
                label_set = set([lbl for lbl in current_labels if lbl != "_"])
                label = list(label_set)[0] if label_set else "O"
                texts.append(full_text)
                labels.append(label)
                current_tokens, current_labels = [], []
            continue

        parts = line.strip().split("\t")
        if len(parts) >= 4:
            token, label = parts[2], parts[3]
            current_tokens.append(token)
            current_labels.append(label)

    return pd.DataFrame({"text": texts, "label": labels})


# 2. Lire tous les fichiers TSV du dossier
tsv_files = glob.glob("tsv/*.tsv")
df_all = pd.concat([parse_webanno_tsv(f) for f in tsv_files], ignore_index=True)
df_all = df_all[df_all['label'] != '']  # sÃ©curitÃ©

# 3. Visualisation en camembert ðŸ°
plt.figure(figsize=(6,6))
df_all['label'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("RÃ©partition des labels")
plt.ylabel('')
plt.tight_layout()
plt.savefig("label_distribution_pie.png")
plt.close()

# 4. Encoder les labels
label_encoder = LabelEncoder()
df_all['label_encoded'] = label_encoder.fit_transform(df_all['label'])

# 5. Split train/test
train_df, test_df = train_test_split(df_all, test_size=0.2, random_state=42)

# 6. Tokenizer BERT
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

def tokenize(batch):
    return tokenizer(batch["text"], padding="max_length", truncation=True)

train_dataset = Dataset.from_pandas(train_df[['text', 'label_encoded']])
test_dataset = Dataset.from_pandas(test_df[['text', 'label_encoded']])
train_dataset = train_dataset.map(tokenize, batched=True)
test_dataset = test_dataset.map(tokenize, batched=True)

train_dataset = train_dataset.rename_column("label_encoded", "labels")
test_dataset = test_dataset.rename_column("label_encoded", "labels")

train_dataset.set_format("torch", columns=["input_ids", "attention_mask", "labels"])
test_dataset.set_format("torch", columns=["input_ids", "attention_mask", "labels"])

# 7. Initialiser le modÃ¨le BERT
model = BertForSequenceClassification.from_pretrained(
    "bert-base-uncased",
    num_labels=len(label_encoder.classes_)
)

# DÃ©placer le modÃ¨le sur le pÃ©riphÃ©rique
model.to(device)

# 8. Fonction de calcul des mÃ©triques
def compute_metrics(p):
    predictions, labels = p
    # VÃ©rifier si predictions est un tensor PyTorch, sinon ne pas appeler .cpu()
    if isinstance(predictions, torch.Tensor):
        predictions = predictions.cpu()  # DÃ©placer les prÃ©dictions sur le CPU si c'est un tensor
    preds = torch.argmax(predictions, axis=1) if isinstance(predictions, torch.Tensor) else np.argmax(predictions, axis=1)

    # Convertir en numpy si ce n'est pas dÃ©jÃ  un tableau numpy
    preds = preds.numpy() if isinstance(preds, torch.Tensor) else preds
    labels = labels.numpy() if isinstance(labels, torch.Tensor) else labels
    accuracy = accuracy_score(labels, preds)
    return {"accuracy": accuracy}



print("Exemples de donnÃ©es avant l'entraÃ®nement :")
print(df_all[['text', 'label']].head())

# 9. EntraÃ®nement
training_args = TrainingArguments(
    output_dir="./bert_results",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    num_train_epochs=50,  # Nombre d'Ã©poques augmentÃ©
    per_device_train_batch_size=25,
    per_device_eval_batch_size=25,
    load_best_model_at_end=True,
    logging_dir="./logs",
    logging_steps=5,
    report_to="none",  # DÃ©sactiver les rapports sur W&B
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    compute_metrics=compute_metrics,  # Ajout des mÃ©triques
)

trainer.train()

def display_predictions(model, tokenizer, texts, label_map):
    model.eval()
    model.to(device)  # DÃ©placer le modÃ¨le sur le bon pÃ©riphÃ©rique
    with torch.no_grad():
        inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt").to(device)
        outputs = model(**inputs)
        predictions = torch.argmax(outputs.logits, dim=1).cpu().numpy()  # DÃ©placer les prÃ©dictions sur le CPU

    for text, pred in zip(texts, predictions):
        print(f"Text: {text}")
        print(f"Predicted label: {label_map[str(pred)]}\n")

# 10. Sauvegarder tokenizer, modÃ¨le et mapping de labels
model.save_pretrained("./bert_trained_model")
tokenizer.save_pretrained("./bert_trained_model")

with open("label_mapping.json", "w") as f:
    json.dump({i: label for i, label in enumerate(label_encoder.classes_)}, f)

# 11. Fonction prÃ©diction ðŸ§™â€â™‚ï¸
def predict(text):
    tokenizer = BertTokenizer.from_pretrained("./bert_trained_model")
    model = BertForSequenceClassification.from_pretrained("./bert_trained_model")
    model.eval()
    model.to(device)

    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True).to(device)
    with torch.no_grad():
        outputs = model(**inputs)
    predicted_class = torch.argmax(outputs.logits, dim=1).item()

    with open("label_mapping.json") as f:
        label_map = json.load(f)

    return label_map[str(predicted_class)]

# Test rapide
print("Exemple : ", predict("Forest fire is spreading fast!"))

# 12. Calcul et visualisation de la matrice de confusion
def plot_confusion_matrix(true_labels, pred_labels):
    cm = confusion_matrix(true_labels, pred_labels)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap="Blues", xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
    plt.title('Matrice de confusion')
    plt.xlabel('PrÃ©dictions')
    plt.ylabel('VÃ©ritables Labels')
    plt.show()

# PrÃ©dictions sur le jeu de test
test_preds = trainer.predict(test_dataset)

# Convertir les prÃ©dictions en tensor PyTorch
test_preds_tensor = torch.tensor(test_preds.predictions)

# Appliquer argmax pour obtenir les indices des classes prÃ©dites
pred_labels = torch.argmax(test_preds_tensor, axis=1).cpu().numpy()

# RÃ©cupÃ©rer les Ã©tiquettes rÃ©elles (s'assurer que c'est un tableau NumPy)
true_labels = test_dataset["labels"].cpu().numpy()


# Affichage de la matrice de confusion
plot_confusion_matrix(true_labels, pred_labels)
```
